```python
import numpy as np
```

```python
x = np.array([[-3.5],
              [-2.6],
              [0.3],
              [1.6]
])
```

```python
y = np.array([[-17.3],
              [-13.8],
              [-1.9],
              [3.8]])
y
```

```python
A = np.concatenate(((np.ones(shape=(4,1))),x),axis=1)
A
```

```python
A.T@A
```

```python
np.linalg.inv(A.T@A)
```

```python
w = np.linalg.inv(A.T@A) @ A.T @ y
w
```

```python
Y_he  = A@w
Y_he 
```

```python
Y_he - y
```

```python
mse  = 1/4 * np.linalg.norm(Y_he - y)**2
mse
```
