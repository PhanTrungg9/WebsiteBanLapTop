```python
import numpy as np
```

```python
B = np.array([[-0.09,-0.09,0.15],
              [-0.07,-0.2,0.13],
              [-0.03,0.18,0.12]
])
```

```python
g = np.array([3,0,-1])
```

```python
q = np.linalg.norm(B,np.inf)
q
```

```python
n = 0
x0 = np.array([1.0,1.0,-1.0])
while n <1000 :
    n+=1
    x = x0.copy()
    for i in range(3):
        x[i] = B[i,:]@x + g[i]
    e  = np.linalg.norm((x-x0),np.inf)
    x0 = x 
    if e <1e-5 :
        print("*",n,x,e)
        break
    print(n,x,e)

```
